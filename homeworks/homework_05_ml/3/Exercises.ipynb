```python
#crime.index=crime.Year
crime=crime.set_index('Year')
crime.head()
```

### Step 7. Удалите колонку Total

```python
crime = crime.drop('Total', axis=1)
crime.head()
```

### Step 8. Сгруппируйте года по декадам и просуммируйте некоторые значения.

#### Осторожнее с Population, суммирование этой колонки - неправильно. Что вместо этого с ней лучше сделать?

```python
crime['decada']=[((crime.index[i].year-crime.index[0].year)//10+1) for i in range(crime.shape[0])]
crime.head(7)
```

```python
dec_crime=crime.groupby('decada').sum()
count_dec=np.array(crime.groupby('decada')['Murder'].count())
population=(np.array(dec_crime.Population)//count_dec)
dec_crime['Population']=pd.Series(population, index=dec_crime.index)
dec_crime
```

### Step 9. Какая самая опасная декада в US?

```python
dec_crime['total']=dec_crime.Violent+dec_crime.Murder+dec_crime.Forcible_Rape+dec_crime.Robbery+dec_crime.Aggravated_assault+dec_crime.Burglary+dec_crime.Larceny_Theft+dec_crime.Vehicle_Theft
dec_crime
```

```python
crime.index[0].year+dec_crime[dec_crime.total==dec_crime.total.max()].index[0]*10-10
```
